 Problem 1 : Payslip Generation
image.png



```python
EName = input("Enter the Name: ")
ENumber = input("Enter the Number: ")
WkEnding = input("Ending Week: ")

HrsWrked = float(input('Enter Number of hours worked: '))
Normal_Hrs = float(37.50)
if HrsWrked > Normal_Hrs:
  Overtime = HrsWrked-Normal_Hrs
  HrsWrked = 37.50
else:
  Overtime = 0

HrsRate = float(input("Enter Hourly Rate: "))
Norm_Total= HrsWrked*HrsRate

OT_Rate= float(input("Enter Overtime Rate: "))
TotalOT_Rate= HrsRate*OT_Rate
OT_Total = round((TotalOT_Rate*Overtime),2)

StdTX=float(input("Enter Standard Tax Rate % : "))
tx1 = round((Norm_Total/StdTX),2)

OT_Tx=float(input("Enter Overtime Tax Rate %: "))
tx2 = round((OT_Total/StdTX),2)

Total_Pay=round((Norm_Total+OT_Total),2)
Total_Deduct=round((tx1+tx2),2)
Net_pay=round((Norm_Total+OT_Total-tx1-tx2),2)


print("                              #Payslip#")
print("WEEK ENDING"+WkEnding)
print("Employee Name: "+EName)
print("Employee Number: "+ENumber)
print("                          Earnings                          Total tax Deductions")
print("                          Hours       Rate         Totals")
print("Hours(normal)              "+str(Normal_Hrs)+"      "+str(HrsRate)+"           "+str(Norm_Total)+"     Tax @ "+str(tx1))
print("Hours(Over-time)            "+str(Overtime)+"      "+str(TotalOT_Rate)+"           "+str(OT_Total)+"     Tax @ "+str(tx2)+"\n")
print("                             Total Pay:                                "+str(Total_Pay))
print("                             Total Deductions:                         "+str(Total_Deduct))
print("                             Net Pay:                                  "+str(Total_Pay))
```

 Problem 3: Frequency of each integer calculation 
 image.png

```python
def CountFrequency(my_list):
    # Creating an empty dictionary
    freq = {}
    for item in my_list:   #For loop for counting the occurance
        if (item in freq):
            freq[item] += 1
        else:
            freq[item] = 1
    print ("\nThe Frequency of all element in the list is mentioned below :")
    for key, value in freq.items():             #for loop for printing the occurances
        print ("%d occurs: %d times"%(key, value))

# Driver function
if __name__ == "__main__":
  my_list=[]
  print('###################################')
  print("WELCOME TO OUR DBS CONSOLE")
  print('###################################')
  n=int(input("Enter the number of elements to be stored in the list :")) #Input
  print("Input ",n," elements in the list:")
  for i in range(0,n):
    e=int(input("Enter Element# - "+str(i)+" : "))
    my_list.append(e)                           #Storing the inputs

CountFrequency(my_list)
```

```python
def welcome():
    #entry variable using the input function
    entry = int(input("""#############################
>>>MYPY Phone Book System<<<
#############################
1. Add New Entry
2. Delete an entry
3. Update Entry
4. Lookup Number
5. Quit\n"""))
    
#Close the function    
    return entry

#Define a function Phonebook
def phonebook():
     #an empty contact dictionary to store
    contact = {}
    
    #while loop to continuously run the phonebook program
    while True:
        #calling welcome function and Set entry variable
        entry = welcome()
        
        if entry == 1:
            #Request for phone number and contact name
            phone_number = input('Please Enter a number: ')
            contact_name = input('Enter in contact name in the format "FirstName,LastName".: ')
            
            #Check if the contact number already exits in Phonebook
            #message to inform the user that contact already exits
            if phone_number in contact.values():
                print("Contact"+ contact_name+ "already exits in your Phonebook")
                
            #If it does not, update current contact list in PhoneBook
            else:
                contact[contact_name]=phone_number
                #Print a success message
                print('Contact successfully saved')
                print('Your updated phonebook: ')
                #each contact in a separate line
                for k, v in contact.items():
                    print(k, '-->', v)
        
        elif entry == 2:
            name = input('Enter the name of the contact you want to delete: ')
            #check if contact exists
            if name in contact:
                #print the required contact
                print('The contact is',name,':',contact[name])

                #Else inform the user that the contact does not exist
            else:
                print('That contact does not exist! Return to the main menu to enter the contact')
                        
            #Confirm if user wishes to delete contact
            confirm = input('Are you sure you wish to delete this contact? Yes/No: ')
            #Initiate a decision
            if confirm.capitalize() == 'Yes':
                #If yes, delete contact from Phonebook
                contact.pop(name,None)
                #Looping through Phonebook and print updated contact list
                for k, v in contact.items():
                    print('Your Updated Phonebook is shown below:')
                    print(k, '-->', v)
            #ELse return to main menu
            else:
                print('Returned to Main Menu')
  
        elif entry == 3:
            contact_name = input('Enter the name of the contact you want to update: ')
            #check if contact exists
            if contact_name in contact:
                #print the required contact
                print('The contact is', contact_name,':',contact[phone_number])
                confirm = input('Do you wish to update contact number? Yes/No: ')
                if confirm.capitalize() == 'Yes':                        
                    #If yes, add contact to Phonebook
                  Temp_phone_number = input('Please Enter a new number: ')
                
                  #Check if the contact number already exits in Phonebook
                  #message to inform the user that contact already exits
                  if Temp_phone_number in contact.items():
                      print('That contact already exits in your Phonebook')
                    
                  #If it does not, update current contact list in PhoneBook
                  else:
                      contact.update({contact_name:Temp_phone_number})
                      #Print a success message
                      print('Contact successfully saved')
                      print('Your updated phonebook: ')
                      #each contact in a separate line
                      for k, v in contact.items():
                        print(k, '-->', v)

                #Else informing the user that the contact does not exist
            else:
                print('That contact with the same name does not exist!')
            

        elif entry == 4:
            name = input('Enter the name of the contact details you wish to view: ')
            #Creating a condition to check if the entered name is in the phonebook
            if name in contact:
                #If yes, print the required contact
                print('The contact is',name,':',contact[name])
                
                #Else inform the user that the contact does not exist
            else:
                print('That contact does not exist!')
            
                confirm = input('Do you wish to add this contact? Yes/No: ')
                if confirm.capitalize() == 'Yes':
                    #If yes, add contact to Phonebook
                  phone_number = input('Please Enter a number of '+name+' :')
                  #contact_name = input('Save it in the format "FirstName,LastName".: ')
                #ELse return to main menu
                else:
                    print('Return to Main Menu')            

                  #Check if the contact number already exits in Phonebook
                  #message to inform the user that contact already exits
                if phone_number in contact.items():
                      print('That contact already exits in your Phonebook')
                    
                  #If it does not, update current contact list in PhoneBook
                else:
                      contact.update({name:phone_number})
                      #Print a success message
                      print('Contact successfully saved')
                      print('Your updated phonebook: ')
                      #each contact in a separate line
                      for k, v in contact.items():
                        print(k, '-->', v)

        elif entry == 5:
            print('Thanks for using the MYPY Phone Book System')
            break
        
        #elif entry == 6:
            #For first entry Check if contact dict is empty
            #If not empty,i.e bool(contact)==False, Print current contact list
            #if bool(contact) != False:
                #for k, v in contact.items():
                    #print(k, '-->', v)
             #Else inform the user that the contact book is empty
            #else:
                #print('You have an empty phonebook! Go back to the menu to add a new contact')            
        #Error Message    
        else:
            print('Incorrect Entry!')
            
    
phonebook()
```
